```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
from eralchemy import render_er

from sqlalchemy import (MetaData, Table, Column, Integer, String, ForeignKey, Boolean)    
metadata = MetaData()

# create your own model ....
users = Table('users', metadata,
    Column('_id', String(40), primary_key=True),
    Column('active', String(15), nullable=False),
    Column('createdDate', Integer(), nullable=False),
    Column('lastLogin', Integer(), nullable=False),
    Column('role', String(15), nullable=False),
    Column('signUpSource', String(15), nullable=False),
    Column('State', String(15), nullable=False)
)    
receipts = Table('receipts', metadata,
    Column('_id',String(40), primary_key=True),
    Column('bonusPointsEarned', Integer()),
    Column('bonusPointsEarnedReason', String(100)),
    Column('createDate', Integer()),
    Column('dateScanned', Integer()),
    Column('finishedDate', Integer()),
    Column('modifyDate', Integer()),
    Column('pointsAwardedDate', Integer()),
    Column('pointsEarned', Integer()),
    Column('purchaseDate', Integer()),
    Column('purchasedItemCount', Integer()),
    Column('rewardsReceiptItemList', ForeignKey('brands.barcode')),
    Column('rewardsReceiptStatus', String(30)),
    Column('totalSpent', Integer()),
    Column('user_id', ForeignKey('users._id')),
)

brands = Table('brands', metadata,
    Column('_id',String(40), primary_key=True),
    Column('barcode', Integer()),
    Column('category', String(100)),
    Column('categoryCode', String()),
    Column('cpg', String()),
    Column('name', String()),
    Column('topBrand', Boolean()),
    Column('brandCode', String())
)      
               
# add your own table ....

# Show ER model from here
filename = './mymodel.jpg'
y = render_er(metadata,filename)
display(y)
```
